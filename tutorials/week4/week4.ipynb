```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Spring Layout

```python
G = nx.read_gml("can_144.gml")

fig, ax = plt.subplots(figsize=(12,12))
ax.set_title("Spring Layout of can_144.gml", fontsize=16)

nx.draw_spring(G, ax=ax, alpha=0.4);
```

# Fruchterman-Reingold layout

Of polbooks.gml

```python
G = nx.read_gml("polbooks.gml")

pos = nx.fruchterman_reingold_layout(G, iterations=100, k=1, threshold=0.001)

fig, ax = plt.subplots(figsize=(12,12))
ax.set_title("Fruchterman-Reingold layout", fontsize=18)

nx.draw_networkx(G, pos=pos, ax=ax, style="dashed", node_size=600,
                 alpha=0.5, color="red", node_color="yellow", node_shape="8");
```

# FM3 Layout

```python
G = nx.read_gml("polbooks.gml")

fig, ax = plt.subplots(figsize=(12,12))
ax.set_title("FM3 layout of polbooks", fontsize=16)

pos = nx.f(G, iterations=100, k=1, threshold=0.001)

nx.draw_spring(G, ax=ax, alpha=0.5)
plt.show()
```

```python
pos = nx.fruchterman_reingold_layout(G, iterations=200, k=3)

fig, ax = plt.subplots(figsize=(12,12))
ax.set_title("Fruchterman-Reingold layout", fontsize=18)

nx.draw_networkx(G, pos=pos, ax=ax, style="dashed", node_size=800,
                 alpha=0.4, color="red", node_color="yellow", node_shape="8");
```
