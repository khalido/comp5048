```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
```

# Florentine

```python
G = nx.florentine_families_graph()
G
```

```python
cc = nx.closeness_centrality(G)
cc
```

```python
core = nx.core_number(G)
core
```

```python
deg = nx.degree(G)
sizes = [1200 * deg[n] for n in G.nodes()]
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.set_title("Florentine Families", fontsize=18)

pos = nx.layout.spring_layout(G, k=1, iterations=100, seed=1)

nx.draw_networkx(G, pos=pos, ax=ax, alpha=0.8, node_size=sizes)
```

# 2. kmeans clustering iris dataset

```python
df = sns.load_dataset("iris")
df.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X=df.iloc[:,0:4])
kmeans.labels_
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X=df.iloc[:,0:4])
clusters = kmeans.labels_
```

```python
sns.scatterplot(x=df.sepal_length, y=df.petal_length, hue=clusters);
```
